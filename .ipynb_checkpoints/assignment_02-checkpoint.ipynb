```python
def Joukowski(z,c):
    
    squigly = z + c**2/z
    
    return squigly
```

```python
c = 1
r = 1.5
theta = numpy.linspace(0,2*math.pi,1000)
z1 = r*numpy.cos(theta) + r*numpy.sin(theta)*1j
pyplot.plot(z1.real,z1.imag)
```

```python
j1 = Joukowski(z1,c)
pyplot.plot(j1.real,j1.imag)
```

```python
r = 1.2
z2 = numpy.cos(theta)-(c-r) + numpy.sin(theta)*1j
```
